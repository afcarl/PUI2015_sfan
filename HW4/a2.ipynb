```python
import pylab as pl
import pandas as pd
import numpy as np
%pylab inline

from scipy import stats
import os

#this makes my plots pretty! but it is totally not mandatory to do it
import json
s = json.load( open("fbb_matplotlibrc.json") )
pl.rcParams.update(s)
```

```python
dist_n = np.random.randn(1000)

threshold = stats.anderson(dist_n, dist='norm')[1][stats.anderson(dist_n, dist='norm')[2]==[1.0]]
```

### Binomial

```python
# use the same mean(50) and step through each probability from 0.1 to 0.9
n = 50
ks_b = np.zeros(9)
ad_b = np.zeros(9)
kl_b = np.zeros(9)
chi2_b = np.zeros(9)


def mynorm (x, mu, var):
    return stats.norm.cdf(x, loc=mu, scale=var)

for i in range(9):
    p=(i+1)/float(10)
    dist = np.random.binomial(n, p, 1000) 
 
    ks_b[i] = stats.kstest(dist, mynorm, args=(n*p, n*p*(1.0-p)))[0]
    ad_b[i] = stats.anderson(dist, dist='norm')[0]
    
    mybins=np.linspace(min(dist),max(dist), 10)
    bincenters = mybins[:-1]+0.5*(mybins[1]-mybins[0])
    kl_b [i] =  stats.entropy(np.histogram(dist, bins=mybins)[0], stats.norm.pdf(bincenters, loc=n*p, scale=n*p*(1.0-p)))
```

```python
fig = pl.figure(figsize = (15,5))
fig.add_subplot(131)
pl.plot(np.arange(0.1,1.0,0.1), ks_b, label='KS')
pl.legend()

fig.add_subplot(132)
pl.plot(np.arange(0.1,1.0,0.1), ad_b,  label='AD')
pl.plot([0.1, 0.9],[threshold, threshold])
pl.legend()

fig.add_subplot(133)
pl.plot(np.arange(0.1,1.0,0.1), kl_b, label='K-L ')
pl.legend()
```

The Anderson-Darling test seems to work the best.

### Poisson

```python
# use the same mean(50) and step through each probability from 0.1 to 0.9
n = 50
ks_b = np.zeros(9)
ad_b = np.zeros(9)
kl_b = np.zeros(9)
chi2_b = np.zeros(9)


def mynorm (x, mu, var):
    return stats.norm.cdf(x, loc=mu, scale=var)

for i in range(9):
    p=(i+1)/float(10) 
    dist = np.random.poisson(n, 1000)
 
    ks_b[i] = stats.kstest(dist, mynorm, args=(n, n))[0]
    ad_b[i] = stats.anderson(dist, dist='norm')[0]
    
    mybins=np.linspace(min(dist),max(dist), 10) 
    bincenters = mybins[:-1]+0.5*(mybins[1]-mybins[0])

    kl_b [i] =  stats.entropy(np.histogram(dist, bins=mybins)[0], stats.norm.pdf(bincenters, loc=n*p, scale=n*p*(1.0-p)))
```

```python
fig = pl.figure(figsize = (15,5))
fig.add_subplot(131)
pl.plot(np.arange(0.1,1.0,0.1), ks_b, label='KS')
pl.legend()

fig.add_subplot(132)
pl.plot(np.arange(0.1,1.0,0.1), ad_b,  label='AD')
pl.plot([0.1, 0.9],[threshold, threshold])
pl.legend()

fig.add_subplot(133)
pl.plot(np.arange(0.1,1.0,0.1), kl_b, label='K-L ')
pl.legend()
```

The K-L test seems to work the best.
