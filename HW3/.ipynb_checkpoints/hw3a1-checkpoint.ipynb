```python
__auther__ = 'sf'

import sys
import pylab as pl
import numpy as np
%pylab inline

import os
import json
s = json.load( open("fbb_matplotlibrc.json" ))
pl.rcParams.update(s)
```

```python
mu = 100
distr = {}
distr['chisq'] = np.random.chisquare(mu,size=100)
```

```python
pl.hist(distr['chisq'], bins = 20)
pl.ylabel('N')
pl.xlabel('x')
```

```python
print 'Chisq — mean: %.2f, standard deviation: %.2f' % (distr['chisq'].mean(),distr['chisq'].std())
```

```python
mysizes = 20000/(np.array(range(1,90))+10)
```

```python
print (mysizes)
```

```python
distr={'chisq':{}}
for n in mysizes:
    distr['chisq'][n] = np.random.chisquare(mu,n)
```

```python
means = {}
means['chisq'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distr['chisq']:
    axchisq_mu_n.plot(nn, distr['chisq'][nn].mean(),'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysizes), max(mysizes)], [mu,mu], 'k')
```
