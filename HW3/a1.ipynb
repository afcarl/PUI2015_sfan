```python
__auther__ = 'sf'

import sys
import pylab as pl
import numpy as np
import random as rd
import scipy.stats as ss
%pylab inline

import os
import json
s = json.load( open("fbb_matplotlibrc.json" ))
pl.rcParams.update(s)
```

### Create an array of 100 random sizes between 10 and 2000
### Standardize the population mean to 100

```python
mu = 100
mysizes = rd.sample(range(10,2000),100)
print mysizes
```

### Set 6 different distributions

```python
distrList = ['normal','poisson','binomial','chisq','triangular','gamma']
distr = {}
for names in distrList:
    distr[names] = {}
print sorted(distr)
```

### Generate samples for 100 different sizes of 6 different distributions

```python
df = 100
for size in mysizes:
    distr['normal'][size] = np.random.normal(mu,10,size)
    distr['poisson'][size] = np.random.poisson(mu,size)
    distr['binomial'][size] = np.random.binomial(mu*2,0.5,size)
    distr['chisq'][size] = np.random.chisquare(df,size)
    distr['triangular'][size] = np.random.triangular(mu-20,mu,mu+20,size)
    distr['gamma'][size] = np.random.gamma(mu,1,size)
```

### Plot all the samples (sample mean vs sample size) for each of the 6 different distributions

```python
meanList = []
for key in distr:
    fig_mu_N = pl.figure(figsize=(6,6))
    ax_mu_n = fig_mu_N.add_subplot(111)
    for n in distr[key]:
        meanList.append(distr[key][n].mean())
        ax_mu_n.plot(n, distr[key][n].mean(),'o')
        ax_mu_n.set_xlabel('sample size')
        ax_mu_n.set_ylabel('sample mean')
        ax_mu_n.set_title(key)
        ax_mu_n.plot([min(mysizes), max(mysizes)], [mu,mu], 'k')
```

### As the sample size gets larger, the sample mean tends to become closer to 100, which is our theoretical mean. It is just as I expected, because the result is supposed to more accurate, or more consistent with the theory, when the sample set gets bigger.

```python
print len(meanList)
n, bins, patches = pl.hist(meanList, bins=100, normed=1)
pl.ylabel('N')
pl.xlabel('sample mean')
pl.title('Histogram of all the sample means')
x = np.linspace(97,103,100)
param = ss.norm.fit(meanList)
y = ss.norm.pdf(bins, param[0], param[1])
pl.plot(bins,y,'r-')
```
